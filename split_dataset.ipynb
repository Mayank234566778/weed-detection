```python
import os
import shutil
import random

# 🔹 Exact source folder
SOURCE_FOLDER = "crop_weed_dataset/images/data"
BASE_DIR = "crop_weed_dataset/images"

print("Looking inside:", SOURCE_FOLDER)
print("Files found:", len(os.listdir(SOURCE_FOLDER)))

# Detect image extensions automatically
images = [f for f in os.listdir(SOURCE_FOLDER)
          if f.lower().endswith((".jpg", ".jpeg", ".png"))]

print("Images detected:", len(images))

if len(images) == 0:
    print("❌ No images detected. Check extension.")
    exit()

random.shuffle(images)

total = len(images)
train_split = int(0.7 * total)
valid_split = int(0.9 * total)

train_files = images[:train_split]
valid_files = images[train_split:valid_split]
test_files = images[valid_split:]

# Create valid and test folders
os.makedirs(os.path.join(BASE_DIR, "valid"), exist_ok=True)
os.makedirs(os.path.join(BASE_DIR, "test"), exist_ok=True)

def move_files(file_list, split_name):
    for img_file in file_list:
        txt_file = os.path.splitext(img_file)[0] + ".txt"

        shutil.move(
            os.path.join(SOURCE_FOLDER, img_file),
            os.path.join(BASE_DIR, split_name, img_file)
        )

        if os.path.exists(os.path.join(SOURCE_FOLDER, txt_file)):
            shutil.move(
                os.path.join(SOURCE_FOLDER, txt_file),
                os.path.join(BASE_DIR, split_name, txt_file)
            )

move_files(valid_files, "valid")
move_files(test_files, "test")

print("✅ Split Completed")
print("Train:", len(train_files))
print("Valid:", len(valid_files))
print("Test:", len(test_files))
```
