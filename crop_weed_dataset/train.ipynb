```python
from ultralytics import YOLO
import torch

print("MPS Available:", torch.backends.mps.is_available())

# Load pretrained small model (best for Mac)
model = YOLO("yolov8n.pt")

# Train
model.train(
    data="weed.yaml",
    epochs=50,
    imgsz=640,
    batch=8,        # reduce to 4 if memory error
    device="mps"    # use "cpu" if MPS not available
)

print("Training Completed ✅")
```

```python
from ultralytics import YOLO

# Load best trained model
model = YOLO("runs/detect/train6/weights/best.pt")

# Run evaluation
metrics = model.val(data="weed.yaml")

print("\n===== Evaluation Results =====")
print(f"Precision: {metrics.box.mp:.4f}")
print(f"Recall: {metrics.box.mr:.4f}")
print(f"mAP50: {metrics.box.map50:.4f}")
print(f"mAP50-95: {metrics.box.map:.4f}")
```

```python
from ultralytics import YOLO
import pandas as pd

# Load model
model = YOLO("runs/detect/train6/weights/best.pt")

# Run validation
metrics = model.val(data="weed.yaml", plots=True)

print("\n===== FINAL EVALUATION =====")
print(f"Precision: {metrics.box.mp:.4f}")
print(f"Recall: {metrics.box.mr:.4f}")
print(f"mAP@0.5: {metrics.box.map50:.4f}")
print(f"mAP@0.5:0.95: {metrics.box.map:.4f}")

# Save metrics
results_dict = {
    "Precision": [metrics.box.mp],
    "Recall": [metrics.box.mr],
    "mAP50": [metrics.box.map50],
    "mAP50-95": [metrics.box.map]
}

df = pd.DataFrame(results_dict)
df.to_csv("final_metrics.csv", index=False)

print("\nMetrics saved to final_metrics.csv")
```

```python
from ultralytics import YOLO
import cv2
import os

model = YOLO("runs/detect/train6/weights/best.pt")

output_folder = "paper_predictions"
os.makedirs(output_folder, exist_ok=True)

results = model.predict(
    source="images/val",
    imgsz=768,       # higher resolution
    conf=0.40,       # cleaner boxes
    save=False       # we will manually save
)

for i, r in enumerate(results):
    img = r.plot(line_width=3)  # thicker bounding boxes
    cv2.imwrite(f"{output_folder}/prediction_{i}.jpg", img)

print("High-quality images saved in 'paper_predictions' folder ✅")
```

```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
import os

# Load trained model
model = YOLO("runs/detect/train6/weights/best.pt")

# Create output folder
output_folder = "paper_figures"
os.makedirs(output_folder, exist_ok=True)

# Run prediction
results = model.predict(
    source="images/valid",      # your validation folder
    imgsz=768,           # higher resolution
    conf=0.45,           # cleaner output
    save=False
)

# Save high-quality images
for i, r in enumerate(results[:6]):   # save first 6 images
    img = r.plot(line_width=5)        # thicker bounding boxes

    plt.figure(figsize=(8, 8))
    plt.imshow(img)
    plt.axis("off")
    plt.tight_layout()

    plt.savefig(
        f"{output_folder}/figure_{i+1}.png",
        dpi=300,                      # publication quality
        bbox_inches="tight",
        pad_inches=0
    )
    plt.close()

print("Publication-quality figures saved in 'paper_figures' folder ✅")
```
